### What Defines a Programming language?

*  Semantics are the meaning of the various symbols used in the language
    *  (e.g. "+", "def", "while", "list")
*  Syntax are the rules used to connect the various symbols
    *  (e.g. "a=5" is syntactically correct in Python but "5=a" is not).

### What are the parts of a computer program?

*  Data:
* Input:
    *  Get data from the keyboard, a file, or some other device.
*  Output:
    *  Display data on the screen or send data to a file or other device.
* Manipulation:
    *  For example, perform basic mathematical operations like addition and multiplication.
*  Conditional execution:
    *    Check for certain conditions and execute the appropriate sequence of statements.
*  Repetition:
    *  Perform some action repeatedly, usually with some variation. (Adapted from *How To Think Like a Computer Scientist: Learning with Python*)


## How Do You Create a Program?
### *Analyze the problem:*
*  Sit down and sketch out the problem.
*  What data will we have, need, manipulate, etc.?
*  What do we want to have when the program is finished executing?

### *Outline the Solution:*
*  Use "pseudo code" (semi-structured natural language)
*  How will program run?
*  How will data be represented?

### *Write the program:*
*  Formal programming language (e.g. Python, Java, C++)
*  *Debug the program:*
    *  Find syntax errors (formal language rules)
    *  Find semantic errors (program runs but doesn't do want you expected).

### *Evaluate the program:*
*  If program isn't satisfactory return to step one and repeat
